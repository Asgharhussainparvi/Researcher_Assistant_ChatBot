```python
# Example: Using references like pointers
a = [10]        # a list is mutable
b = a           # b references the same list as a

b[0] = 99       # modifying through b affects a
print(a) 
```

```python
a = 10
c = id(a)   # get the memory address of the object
print(c)    # prints the integer's memory address
```

```python
b = "Ali"
d = 10
print(id(b)) 
print(id(d)) # should be the same as id(a) since integers are immutable and cached
```

```python
class Pointer:
    def __init__(self, value):
        self.value = value

a = Pointer(10)
b = a  # b points to the same object as a

b.value = 99
print(a.value)  # Output: 99
```
